# Clean the Diabetes Dataset

import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("f:/4TH SEM/Group Project/Code/data/diabetes.csv")
```

To know about the meta data of set like rows and columns

```python
print("Dataset shape:", df.shape)
df.head()
```

To get the overview of dataset

```python
df.info()
df.describe()
```

To find if there are missing value or not

```python
df.isnull().sum()
```

# if missing value exists then run below step

For numerical value we run below code

```python
num_cols = ["age", "bmi", "HbA1c_level", "blood_glucose_level"]

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

For Categorical value we run below code

```python
cat_cols = ["gender", "smoking_history"]

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

We Validate the medical values for the columns 

```python
df = df[df["bmi"] > 0]
df = df[df["age"] > 0]
df = df[df["blood_glucose_level"] > 0] 
```

To encode the categorical columns into number forms as it is necessary during model training

```python
df = pd.get_dummies(
    df,
    columns=["gender", "smoking_history"],
    drop_first=True
)
```

# To Save The Clean Dataset

```python
df.to_csv("c:/4TH SEM/Group Project/Code/data/cleaned_diabetes.csv", index=False)
```
